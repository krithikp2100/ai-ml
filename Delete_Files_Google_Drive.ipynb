```python
import os

# Define the path to your main Google Drive folder
main_drive_path = '/content/drive/My Drive/'

# --- Part 1: Find and LIST files for verification ---
# It is highly recommended to run this part first to see what will be deleted.

print("--- Verifying Files to be Deleted ---")
files_to_delete = []
for filename in os.listdir(main_drive_path):
  # Create the full path to the item
  full_path = os.path.join(main_drive_path, filename)

  # Check if it's a file (not a folder) and if its name starts with "Film"
  if os.path.isfile(full_path) and filename.startswith("Film"):
    files_to_delete.append(filename)
    print(f"Found: {filename}")

print(f"\nFound a total of {len(files_to_delete)} files to delete.")
print("Please review the list above carefully.")
```

```python
# This action is PERMANENT. Proceed only if you are sure.

if files_to_delete: # Only ask for confirmation if files were found
  # Ask for explicit confirmation from the user
  confirmation = input("\nType the word 'DELETE' to permanently remove these files: ")

  if confirmation == "DELETE":
    print("\n--- Starting Deletion ---")
    for filename in files_to_delete:
      try:
        file_path_to_delete = os.path.join(main_drive_path, filename)
        os.remove(file_path_to_delete)
        print(f"Deleted: {filename}")
      except Exception as e:
        print(f"Error deleting {filename}: {e}")
    print("\n✅ Deletion complete.")
  else:
    print("\n❌ Deletion cancelled.")
else:
  print("\nNo files starting with 'Film' were found to delete.")
```
